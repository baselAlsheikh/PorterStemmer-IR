### Porter Stemmer ###

```python
import re

class PorterStemmer:
    def __init__(self):
        self.step1a_rules = [
            (r"sses$", "ss"),
            (r"ies$", "i"),
            (r"ss$", "ss"),
            (r"s$", "")
        ]

        self.step1b_rules = [
            (r"eed$", "ee"),
            (r"ed$", ""),
            (r"ing$", "")
        ]

        self.step2_rules = [
            (r"ational$", "ate"),
            (r"tional$", "tion"),
            (r"enci$", "ence"),
            (r"anci$", "ance"),
            (r"izer$", "ize"),
            (r"abli$", "able"),
            (r"alli$", "al"),
            (r"entli$", "ent"),
            (r"ousli$", "ous"),
            (r"ization$", "ize"),
            (r"ation$", "ate"),
            (r"ator$", "ate"),
            (r"alism$", "al"),
            (r"iveness$", "ive"),
            (r"fulness$", "ful"),
            (r"ousness$", "ous"),
            (r"aliti$", "al"),
            (r"iviti$", "ive"),
            (r"biliti$", "ble"),
        ]

    def measure(self, stem):
        """Approximation of m (VC measure)"""
        form = re.sub(r'[^aeiou]+', 'C', stem)
        form = re.sub(r'[aeiou]+', 'V', form)
        return form.count("VC")

    def stem(self, word):
        word = word.lower()

        # Step 1a
        for pattern, repl in self.step1a_rules:
            word = re.sub(pattern, repl, word)

        # Step 1b
        for pattern, repl in self.step1b_rules:
            if re.search(pattern, word):
                stem = re.sub(pattern, "", word)
                if re.search(r"[aeiou]", stem):  # check if stem has a vowel
                    word = re.sub(pattern, repl, word)
                break

        # Step 2
        for pattern, repl in self.step2_rules:
            word = re.sub(pattern, repl, word)

        # Step 3 (m > 1 condition)
        if re.search(r"ement$", word):
            stem = re.sub(r"ement$", "", word)
            if self.measure(stem) > 1:
                word = stem

        return word


# Example usage
if __name__ == "__main__":
    stemmer = PorterStemmer()
    words = ["replacement", "cement", "relational", "conditional", "ties", "ponies", "caresses"]
    print("Original â†’ Stemmed")
    for w in words:
        print(f"{w:15} â†’ {stemmer.stem(w)}")
```

### ðŸ” Porter Stemmer Explanation

The Porter Stemmer reduces English words to their root form using a sequence of suffix-removal rules.  
It follows the lecture description â€” *â€œfive phases of reductions applied sequentiallyâ€* â€” to normalize words for indexing or search.

**Main Steps:**
1. **Step 1a:** Removes plural endings (`sses â†’ ss`, `ies â†’ i`, `s â†’ ""`).  
2. **Step 1b:** Removes past/continuous forms (`ed`, `ing`) if a vowel remains in the stem.  
3. **Step 2:** Handles complex endings like `ational â†’ ate`, `tional â†’ tion`, `fulness â†’ ful`, etc.  
4. **Step 3 (m>1 rule):** Removes `ement` only if the stem is long enough (`replacement â†’ replac`, `cement â†’ cement`).  
5. **Measure Function (m):** Counts vowelâ€“consonant patterns (VC) to prevent over-stemming.

**Goal:** Group similar words (e.g., *connect, connected, connecting â†’ connect*) to improve information retrieval accuracy and efficiency.
